02_train_unet_classifier

Train a U-Net on the patch dataset.


```
from pathlib import Path
from unetloopdetection.config import TrainConfig

cfg = TrainConfig(
    train_npz="../data/train_patches.npz",
    val_npz="../data/val_patches.npz",
    run_dir="../outputs/run",
    epochs=5,
    batch_size=16,
    lr=1e-3,
    device="cuda",
)
Path(cfg.run_dir).mkdir(parents=True, exist_ok=True)
print(cfg)
```

```
from unetloopdetection.train import main as train_main

best_path = train_main(cfg)
print("best checkpoint:", best_path)
```
