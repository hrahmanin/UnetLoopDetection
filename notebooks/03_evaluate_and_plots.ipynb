03_evaluate_and_plots

Load a checkpoint, run evaluation, and plot a few predictions.


```
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

from unetloopdetection.eval import load_model
from unetloopdetection.dataset import PatchDataset

ckpt = Path("../outputs/run/model_best.pt")
ds = PatchDataset("../data/val_patches.npz", has_labels=True)

device = "cuda" if __import__("torch").cuda.is_available() else "cpu"
model, cfg = load_model(ckpt, device=device)
print("loaded:", ckpt)
print("device:", device)
```

```
import torch

# take a few examples
n_show = 4
x, y = [], []
for i in range(n_show):
    xi, yi = ds[i]
    x.append(xi)
    y.append(yi)
x = torch.stack(x, dim=0).to(device)
y = torch.stack(y, dim=0).cpu().numpy()

with torch.no_grad():
    p = torch.sigmoid(model(x)).cpu().numpy()

print("pred:", p.shape)
```

```
# plots
for i in range(n_show):
    fig = plt.figure(figsize=(9,3))
    ax1 = fig.add_subplot(1,3,1)
    ax2 = fig.add_subplot(1,3,2)
    ax3 = fig.add_subplot(1,3,3)

    ax1.imshow(x[i,0].detach().cpu().numpy())
    ax1.set_title("X")
    ax1.axis("off")

    ax2.imshow(y[i,0])
    ax2.set_title("y")
    ax2.axis("off")

    ax3.imshow(p[i,0])
    ax3.set_title("p")
    ax3.axis("off")

    plt.show()
```
