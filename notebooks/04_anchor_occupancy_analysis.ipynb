```
import numpy as np
import pandas as pd
from pathlib import Path

# Example: load patch-level probabilities saved by scripts/score_pairs.py
score_path = Path("../outputs/scores.npz")
if not score_path.exists():
    print("No scores found at", score_path)
    print("Run: python ../scripts/score_pairs.py --ckpt ../outputs/run/model_best.pt --npz ../data/val_patches.npz --out ../outputs/scores.npz")
else:
    z = np.load(score_path)
    probs = z["probs"]  # (N, 1, H, W)
    print("probs:", probs.shape)
```

```
# Simple 'occupancy' per patch = mean probability
if 'probs' in locals():
    occ = probs.mean(axis=(1,2,3))
    df = pd.DataFrame({"patch_id": np.arange(len(occ)), "occupancy": occ})
    df.describe()
```

```
# Save table
if 'probs' in locals():
    out = Path("../outputs/anchor_occupancy.csv")
    df.to_csv(out, index=False)
    print("saved:", out)
```
